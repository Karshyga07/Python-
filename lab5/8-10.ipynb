```python
import re
with open("row.txt", 'r', encoding = "utf-8") as f:
    data = f.read()
print("task 9")
pettern = r'[A-Z][^a-z]*'
for match in data.split('\n'):
    matches = re.findall(pettern, match)
    if matches:
        print(f'{match} = {matches}')
```
